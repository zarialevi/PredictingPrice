```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
data = pd.read_csv('query.csv')
data1 = pd.read_csv('query(1).csv')
data2 = pd.read_csv('query(2).csv')
all_data = pd.concat([data,data1,data2])
df = pd.DataFrame(all_data)
```

```python
df.drop('nst', axis=1, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
int_cols = df[['latitude','longitude','depth','mag','gap','dmin','rms','horizontalError','depthError','magError','magNst']]
```

```python
sns.heatmap(int_cols, annot=True)
```

```python
sns.heatmap(int_cols, annot=True, annot_kws={"size": 7})
sns.plt.show()
```
